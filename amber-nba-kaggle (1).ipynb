```python
path = "/kaggle/input/basketball/nba.sqlite"
conn = sql.connect(path)
print("database connected")
```

```python
t1 = pd.read_sql_query("Select * from player",conn)
print(t1)
```

```python
q1 = """
select season, count(distinct team_id) as count_teams
from draft_history
group by season
order by count_teams limit 5;
"""

r1 = pd.read_sql(q1,conn)
print(r1)
```

```python
# where the players are coming from?

q2 = """
select team_name, count(distinct person_id) as total_drafter
from draft_history
group by team_name
order by total_drafter
limit 15;
"""

r2 = pd.read_sql(q2, conn)
print(r2)
```

```python
#what percentage of matches won by home team?

q3 = """
select (count(*)*100.0)/(select count(*) from game) as win_percentage
from game
where wl_home = "W";
"""

r3 = pd.read_sql(q3, conn)
print(r3)
```

```python
#which team has had the most inactive players?

q4 = """
select team_id,
count(*) as inactive_players_count
, rank() over (order by count(*) desc) as rnk
from inactive_players
group by team_id
limit 1
"""

r4 = pd.read_sql(q4, conn)
print(r4)
```
